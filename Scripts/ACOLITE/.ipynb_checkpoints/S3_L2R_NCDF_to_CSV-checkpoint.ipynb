```python
import os
import netCDF4
import numpy as np
import geopandas as gpd
import pandas as pd
import xarray as xr
import gc
```

```python
# set path
path = 'C:/Users/ssharp/Documents/Git/LE2022/Data/Sentinel3/S3_L2_acolite/2020_06_02'
os.chdir(path)
# set filename
ncs = 'S3A_OLCI_2020_06_02_15_46_02_L2R_projected.nc'
# read in netcdf with xarray
ds = xr.open_dataset(ncs)
# convert to pandas dataframe 
edgar = ds.to_dataframe()
# the index in the df is a Pandas.MultiIndex. To reset it, use df.reset_index()
edgar = edgar.reset_index()
# use geopandas points_from_xy() to transform Longitude and Latitude into a list of shapely.Point objects 
# and set it as a geometry while creating the GeoDataFrame
edgar_gdf = gpd.GeoDataFrame(edgar, geometry=gpd.points_from_xy(edgar.lon, edgar.lat))
# Save
edgar_gdf.to_csv("S3A_OLCI_2020_06_02_15_46_02_L2R_projected.csv")
```
